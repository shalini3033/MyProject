```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
```

```python
nRowsRead = 3000        # specify 'None' if want to read whole file
dataset = pd.read_csv('Cleaned-Data.csv',  nrows = nRowsRead)           # delimiter=',',
dataset.dataframeName = 'Cleaned-Data.csv'
nRow, nCol = dataset.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
X = dataset.iloc[:, 0:10].values
y = dataset.iloc[:, 20].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 10, metric = 'minkowski', p = 2)
model =classifier.fit(X_train, y_train)
```

```python
prediction =model.predict(X_test)
print(prediction)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, prediction)
cm
print(metrics.accuracy_score(y_test, prediction))
# accuracy_score(y_test, prediction)
```

```python
# try K=1 through K=30 and record testing accuracy
k_range = list(range(1, 30))
scores = []
for k in k_range:
    classifier = KNeighborsClassifier(n_neighbors = k)
    model =classifier.fit(X_train, y_train)
    prediction =model.predict(X_test)
    scores.append(metrics.accuracy_score(y_test,prediction))
```

```python
# import Matplotlib (scientific plotting library)
import matplotlib.pyplot as plt

# allow plots to appear within the notebook
%matplotlib inline

# plot the relationship between K and testing accuracy
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```
